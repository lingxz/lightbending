```python
from sympy import *
from gravipy import *
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
t, r, phi, tau= symbols('t, r, \\phi, \\tau')
```

```python
x = Coordinates('\chi', [t, r, phi])
```

```python
R = Function('R')(r, t)
M, Lambda = symbols('M, \\Lambda')
# E = Function('E')(r)
a = Function('a')(t)
k = Symbol('k')
alpha = Function('\\alpha')(r, t)
Metric = diag(-1, sqrt(2*M/R + Lambda/3*R**2), R**2)
g = MetricTensor('g', x, Metric)
Ga = Christoffel('Ga', g)
Ri = Ricci('Ri', g)
G = Einstein('G', Ri)
w = Geodesic('w', g, tau)
```
